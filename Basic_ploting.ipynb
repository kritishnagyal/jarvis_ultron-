```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#imporitng the dataframe
d1=pd.read_csv('world-happiness.csv')
d1
```

```python
#lets take two attributes and plot them
#let we take attribute "lowerwhisker" and "Freedom to make life choices" from above dataframe and plot them
lw = d1['lowerwhisker']
FMLC= d1['Freedom to make life choices']
plt.plot(lw,FMLC,color='green',linewidth=0.5)
plt.title('Graph')
plt.xlabel('lower whisker')
plt.ylabel('FMLC')
plt.legend('lw')
#plt.plot(score,expectancy,'colour=green')
```

```python
#for multiple plots
#condition to do this is that the x-axis must be same for both the plots
pc=d1['Perceptions of corruption']
gn=d1['Generosity']
plt.plot(lw,FMLC,color='black',label='Freedom to make life choices')
plt.plot(lw,pc,color='red',label='Perceptions of corruption')
plt.plot(lw,gn,color='purple',label='Generosity')
plt.legend(loc='upper right')
```

```python
#for bar graph
#the attributes must be integer so convert it to the integr
LW = lw.apply(int)
count=LW.value_counts()
lower=count.index
plt.bar(lower,count)
plt.title('Bar graph')
plt.ylabel('lowerwhisker')
plt.xlabel('Indexing')
```

```python
#histogram
plt.hist(lw)
plt.title('Histogram')
plt.xlabel('Indexind')
plt.ylabel('lowerwhisker')
```

```python
#boxplot
plt.boxplot(d1['lowerwhisker'])
plt.title('Box Plot of lowerwhisker')
```

```python
#scatter plt
plt.scatter(pc,lw)
plt.title('scatter plot')
plt.xlabel('Perceptions of corruption')
plt.ylabel('lowerwhisker')
```

```python
#to save the figure 
fig=plt.figure()
plt.hist(lw)
plt.title('Histogram of lowerwhisker')
plt.xlabel('Indixing')
plt.ylabel('lowerscore(lw)')
fig.savefig('Hist_plot.png')
```
